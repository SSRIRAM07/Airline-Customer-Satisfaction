# Data visualization

```python
sns.set_context("poster",font_scale= .7)
plt.figure(figsize=(20, 6))
ax= df["satisfaction"].value_counts().plot(kind="bar",color="crimson",rot=0)
for p in ax.patches:
    ax.annotate(int(p.get_height()),(p.get_x()+0.25,p.get_height()-100),
                ha="center",va="bottom",color="white")
ax.set_xticklabels(["Satisfied","Dissatified"],rotation=0)
plt.show()
```

```python
sns.set_context("poster", font_scale=0.7)
plt.figure(figsize=(20, 6))

# Getting the value counts for the 'satisfaction' column
satisfaction_counts = df["satisfaction"].value_counts()

# Labels for the pie chart
labels = ["Satisfied", "Dissatisfied"]

# Colors for the pie chart
colors = ["crimson", "grey"]

# Exploding the pie slices (adjust the values as needed)
explode = [0.1, 0]  # This will "explode" each slice by 10%

# Plotting the pie chart
wedges, texts, autotexts = plt.pie(
    satisfaction_counts,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct=lambda pct: int(round(pct * satisfaction_counts.sum() / 100)),
    startangle=90,
    counterclock=False,
    textprops={'color':"white", 'fontsize':12}
)

# Adjusting the size of the labels inside the pie chart
for text in autotexts:
    text.set_fontsize(12)

# Setting aspect ratio to be equal to ensure the pie is drawn as a circle
plt.axis('equal')

plt.show()
```

# Model building

```python
# Importing necessary libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
#removing the unwanted columns
df=df.drop(columns="Type of Travel")
```

```python
df=df.drop(columns="Arrival Delay in Minutes")
```

```python
#split the dataset
X=df.drop(columns="satisfaction")
y=df["satisfaction"]
```

```python
X.info()
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
import category_encoders as ce
```

```python
encoder = ce.LeaveOneOutEncoder()
x=encoder.fit_transform(X,y)
```

```python
x.info()
```

# Train Test Spliting

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=40)
```

```python
# Decision Tree
decision_tree = DecisionTreeClassifier()

# Random Forest
random_forest = RandomForestClassifier()

# K-Nearest Neighbors
knn = KNeighborsClassifier()

# Support Vector Machine
svc = SVC()
```

```python
models=[decision_tree,random_forest,knn,svc]
```

# Accuracy of a model

```python
for model in models:
    model_name = model.__class__.__name__
    print(f"Fitting model: {model_name}")
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name} accuracy: {accuracy}")
```
